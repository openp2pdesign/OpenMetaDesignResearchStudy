```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import seaborn as sns
import shutil

pd.set_option('display.max_columns', None) # Display all the columns

sns.set()
#sns.set_style("whitegrid") # Seaborn style
```

```python
data = pd.read_csv("csv/results.csv")
```

```python
columns = ['Q13[SQ001]','Q13[SQ002]','Q13[SQ003]','Q13[SQ004]','Q13[SQ005]','Q13[SQ006]']
columns_names = [
    'My ambition with making is commercially oriented',
    'My ambition with making is socially oriented',
    'My ambition with making is policy oriented',
    'My ambition with making is research oriented',
    'My ambition with making is local community oriented',
    'My ambition with making is technically oriented'
]
# Shorter names for the chart
columns_names = [
    'Commercially oriented ambition',
    'Socially oriented ambition',
    'Policy oriented ambition',
    'Research oriented ambition',
    'Local community oriented ambition',
    'Technically oriented ambition'
]

options = [
    '1 - Strongly disagree',
    '2 - Disagree',
    '3 - Neither agree nor disagree',
    '4 - Agree',
    '5 - Strongly agree',
    'No answer',
]
profile = data[columns]
```

```python
#All subquestions in the same plot
fig = plt.figure(figsize=(15,10))
fig.subplots_adjust(hspace=0.4, wspace=0.4)
fig.suptitle('Q13. What is your ambition with respect to the Maker Movement?', fontsize=22, fontname='Fira Sans', fontweight="500")
axes = []
for k, i in enumerate(columns):
    if k == 0:
        ax = fig.add_subplot(2, 3, k+1)
        axes.append(ax)
    else:
        ax = fig.add_subplot(2, 3, k+1, sharey=axes[0])
        axes.append(ax)
    ax.set_title(columns_names[k], fontsize=14, y=1.02, fontname='Fira Sans', fontweight="500")
    this_profile = profile[i].str.strip().fillna(value="No answer").value_counts(dropna=False).reindex(options).fillna(value=0)
    my_colors = sns.color_palette("husl", len(this_profile)) # Set color palette
    this_profile.plot(kind='barh', color=my_colors, alpha=0.75, rot=0)
    ax.set_xlim([0, 5])
    perc = profile[i].str.strip().fillna(value="No answer").value_counts(normalize=True, dropna=False) * 100
    perc = perc.reindex(options).fillna(value=0)
    for l, v in enumerate(this_profile):
        ax.text(v + .1, l-0.1, str(v)+" ("+str(perc[l])+"%)", fontweight='500', fontsize=9)
        
plt.savefig("svg/Q13.svg", bbox_inches="tight")
plt.savefig("png/Q13.png", dpi=300, bbox_inches="tight")
```
